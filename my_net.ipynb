```python
import numpy as np
x_train = np.random.random((100000, 3))
# print (x_train)
import cv2 as cv

import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense


tr = './true/0.jpg'
fl = './false/1.jpg'

from keras.models import load_model

# img = cv.imread(tr, cv.IMREAD_COLOR)
# img1 = cv.imread(fl, cv.IMREAD_COLOR)
# height, width, channels = img.shape
# print(height)
# print(width)

# height, width, channels = img.shape

rez1 = []
i = 0 
t = 1
while( i < 1):
  img = cv.imread('./'+str(4)+'.jpg', cv.IMREAD_COLOR)
  height, width, channels = img.shape
  g = img.reshape(height*width,3)
  c = np.resize(g,(height*width,3))
  rez1.append(g)
  t = t * (height*width)
  i = i + 1
rez1 = np.array(rez1)
rez1 = np.resize(rez1,(t,3))

rez2 = []
i = 0 
t = 1

while( i < 1):
  img = cv.imread('./'+str(4)+'_.jpg', cv.IMREAD_COLOR)
  height, width, channels = img.shape
  g = img.reshape(height*width,3)
  c = np.resize(g,(height*width,3))
  rez2.append(g)
  t = t * (height*width)
  i = i + 1
rez2 = np.array(rez2)
rez2 = np.resize(rez2,(t,3))


# print(rez2)



model = Sequential()
model.add(Dense(units=1, input_shape=[3]))
model.add(Dense(units=2228,activation='relu'))
model.add(Dense(units=1142,activation='relu'))
model.add(Dense(units=856,activation='relu'))
model.add(Dense(units=241,activation='relu'))
model.add(Dense(units=14,activation='relu'))
model.add(Dense(units=3,))
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array(
    # [[233, 11, 0], [76, 81, 146],[65, 1, 255]]
    rez1
    , dtype=float)
ys = np.array(
    # [[89, 60, 11], [45, 51, 11],[67, 90, 255]]
    rez2
    , dtype=float)

model.fit([xs/255], [ys/255], epochs=15)
# l = model.predict(np.array([[78,45,88]])/255)
model.save('my_model_2.h5')
```
